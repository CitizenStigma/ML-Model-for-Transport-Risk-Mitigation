# Column Transformations

```python
left.dtypes
```

```python
import datetime
from math import sin, cos, pi
```

```python
left.columns
```

```python
def suncalc(s: float):
    if (s < 1):
        return sin(s)
    else:
        return 1 - s
```

```python
left["sun"] = (left.dt - left.sunrise).apply(lambda dt: dt.total_seconds()) / (left.sunset - left.sunrise).apply(lambda dt: dt.total_seconds())
left["dtt"] = left.dt.apply(lambda dt: datetime.datetime(dt.year, dt.month, dt.day, dt.hour, (dt.minute // 5) * 5))
left.drop(columns=["pkId", "fkLinkSerialId", "acceleration", "heading", "direction", "weatherid", "place_type", "accuracy", "place_name", "relevance", "text", "place", "region",
           "locality", "neighborhood", "postcode", "country", "fkReadingsId", "fkSensorSerialId"] + parse_dates, inplace=True)
left.sun = left.sun.apply(suncalc)
left.rename(columns={"dtt": "dt"}, inplace=True)
left
```

```python
right = right.loc[right.drop(columns=["fkReadingsId", "fkSensorSerialId", "measurement", "value"]).dropna(how='all').index]
```

```python
right["sun"] = (right.dt - right.sunrise).apply(lambda dt: dt.total_seconds()) / (right.sunset - right.sunrise).apply(lambda dt: dt.total_seconds())
right["dtt"] = right.dt.apply(lambda dt: datetime.datetime(dt.year, dt.month, dt.day, dt.hour, (dt.minute // 5) * 5))
right.drop(columns=["pkId", "fkLinkSerialId", "acceleration", "heading", "direction", "weatherid", "place_type", "accuracy", "place_name", "relevance", "text", "place", "region",
           "locality", "neighborhood", "postcode", "country", "fkReadingsId", "fkSensorSerialId"] + parse_dates, inplace=True)
right.sun = right.sun.apply(suncalc)
right.rename(columns={"dtt": "dt"}, inplace=True)
right
```

```python
all = pd.concat([left, right], ignore_index=True).dropna(axis=1, how='all')
all = all[~all.value.isna()]
all
```

```python
def sindeg(d: float):
    return sin(d / 180 * pi)
def cosdeg(d: float):
    return cos(d / 180 * pi)
```

```python
all["x"] = all.lat.apply(cosdeg) * all.long.apply(cosdeg)
all["y"] = all.lat.apply(cosdeg) * all.long.apply(sindeg)
all["z"] = all.lat.apply(sindeg)
all.drop(columns=["lat", "long"], inplace=True)
all
```
